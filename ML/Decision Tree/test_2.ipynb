```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from DT import decision_tree
from sklearn.model_selection import train_test_split
```

```python
# data
data = load_breast_cancer()
X,y = data.data,data.target

# train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.6,random_state=42)

n,m = X_train.shape
```

```python
model = decision_tree(max_depth=3,min_leaf_size=2)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
accuracy = model.accuracy_score(y_pred,y_test)
```

```python
print(model)
```

```python
print('Accuracy Score :',accuracy)
```

```python
feat_used = model.get_used_features()
print(f'Features used : {feat_used}')
print(f'\nThus, {len(feat_used)} features used out of available {m}.')
```
