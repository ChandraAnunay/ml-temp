```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as r
from pca import pca
```

Example 1 : dimensionality of original data = 2

```python
# generating 2d data for PCA
n = 80                      # number of data points
mu = np.array([1,3])
sigma = np.array([[1,0.8],[0.8,1]])
data = r.multivariate_normal(mu,sigma,n)
```

```python
# visualizing data
plt.figure(figsize = (5,4))
plt.scatter(data[:,0],data[:,1],marker = '.')
plt.title('Data')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# testing the pca algorithm on the sample data

pca1 = pca.PCA(1)
pca1.fit(data)
principle_components, projection = pca1.transform(data)
```

```python
# visualising the projected data

plt.figure(figsize = (15,4))
plt.subplot(1,2,1)
plt.scatter((data)[:,0],(data)[:,1],marker = '.')
plt.title('Original Data')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.subplot(1,2,2)
plt.scatter(projection[:,0],projection[:,1],marker = '.')
plt.title('Projected space')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# visualising the projected data II
plt.figure(figsize = (8,8))
plt.scatter(data[:,0],data[:,1],marker = '.',label = 'Centralized Data Points')
plt.scatter(projection[:,0],projection[:,1],marker = '.',c='red',label = 'Projected Points')
for i in range(len(data)):
    plt.plot([data[i,0],projection[i,0]],[data[i,1],projection[i,1]],linestyle = 'dotted',c='black')
plt.title('Projecting onto reduced feature space')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
```
