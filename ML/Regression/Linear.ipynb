Given an overdetermined system $Ax=b$, we can determine the coefficients $x$ as below :

$$ x = (A^{T}A)^{-1}A^{T}b $$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data 1

# data_x = np.array([2,4,6,8])
# data_y = np.array([3,7,5,10])

# data 2

data_x = np.array([5,7,12,16,20])
data_y = np.array([40,120,180,210,240])
data = np.array([[data_x[i],data_y[i]] for i in range(len(data_x))])
```

```python
# fitting function 1
def linear_fit_A_b(A,b):
    A_t = np.transpose(A)
    B = np.linalg.inv(np.matmul(A_t,A))
    x = np.matmul(np.matmul(B,A_t),b)
    return x[0],x[1]

# fitting function 2
def linear_fit_xy(data):
    l = len(data)
    A = np.zeros((l,2),float)
    b = data[:,1]
    for i in range(l):
        A[i][0] = data[i,0]
        A[i][1] = 1
    return linear_fit_A_b(A,b)
```

```python
# visualising data

m,c = linear_fit_xy(data)
print('Slope :',m)
print('Intercept :',c)
l,u = min(data_x),max(data_x)

plt.scatter(data_x,data_y,label = 'data')
plt.plot([l,u],[m*l+c,m*u+c],c='red',label = f'fitted line : {m:.3f}x+{c:.3f}')
plt.legend()
plt.show()
```
